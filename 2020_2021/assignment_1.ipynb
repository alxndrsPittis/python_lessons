### Άσκηση 6

Το [δέκαθλο](https://en.wikipedia.org/wiki/Decathlon) είναι ένα ολυμπιακό άθλημα όπου οι αθλητές αγωνίζονται σε 10 διαφορετικά αθλήματα. Η βαθμολογία των αθλητών βασίζεται στον [παρακάτω πίνακα](https://en.wikipedia.org/wiki/Decathlon#Points_system):

|Event | A | B | C |
|------|---|---|---|
|100 m |25.4347 | 18 | 1.81|
|Long jump |0.14354 |220 | 1.4|
|Shot put | 51.39 | 1.5 | 1.05 |
|High jump |0.8465 | 75 | 1.42 |
|400 m | 1.53775 | 82 | 1.81 |
|110 m hurdles | 5.74352 | 28.5  | 1.92|
|Discus throw | 12.91 | 4 | 1.1 |
|Pole vault | 0.2797 | 100 | 1.35 |
|Javelin throw |10.14 | 7 | 1.08 |
|1500 m | 0.03768 | 480 | 1.85 |

Στα αγωνίσματα όπου το μεγαλύτερο σημαίνει καλύτερο (Long jump, Shot put, High jump, Discus throw, Pole vault, Javelin throw), κάθε παίκτης παίρνει πόντους σύμφωνα με τον τύπο:  INT(A(P — B)<sup>C</sup>). Όπου P είναι η επίδωση του αθλητή και A,B,C είναι οι τιμές του πίνακα στο αντίστοιχο άθλημα. Για παράδειγμα ένας παίκτης που κάνει 800cm στο Long Jump, παίρνει: ```int(0.14354 * ((800-220)**1.4))``` = 1061 πόντους.

Ομοίως στα υπόλοιπα αγωνίσματα (αποστάσεις) όπου το μικρότερο είναι το καλύτερο, ο τύπος είναι: INT(A(B — P)<sup>C</sup>). Δηλαδή αν ένας παίκτης στα 100m κάνει 10sec τότε παίρνει: ```int(25.4347*((18-10)**1.81))``` = 1096 πόντους.

Ας υποθέσουμε ότι η μεταβλητή με το όνομα scores περιέχει μία λίστα με τις επιδόσεις ενός αθλητή στα αθλήματα με τη σειρά που υπάρχουν και στον πίνακα. Γράψτε κώδικα python ο οποίος θα υπολογιζει τη συνολική βαθμολογία (άθροισμα πόντων) του παίκτη.

Για τη διευκόλυνσή σας δίνεται ο πίνακας σε μορφή python:

