Έστω οι 5 παρακάτω ακολουθίες DNA:
```
s1 = 'AGTCAATA'
s2 = 'GTC'

```

Παρατηρούμε ότι η ακολουθία ```s2``` μπορεί να γίνει align πάνω στην ακολουθία ```s1```:
```
AGTCAATA
 GTC
```

Επίσης παρατηρούμε ότι η ακολουθία ```s3``` μπορεί να γίνει align πάνω στην ```s1``` αν επιτρέψουμε ένα κενό:
```
AGTCAATA
   CAA A
```

Επίσης παρατηρούμε ότι η ακολουθία ```s4``` μπορεί να γίνει align πάνω στην ```s1``` αν επιτρέψουμε 3 κενά:
```
AGTCAATA
A T A T
```

Τέλος παρατηρούμε ότι η ακολουθία ```s5``` ΔΕΝ μπορεί να γίνει align πάνω στην ```s1``` με κανένα τρόπο. 

Φτιάξτε μία συνάρτηση με το όνομα ```align``` η οποία θα παίρνει ως παράμετρους 2 strings. Η συνάρτηση θα επιτρέφει:
* 0 αν μπορεί να γίνει τέλεια το alignment (π.χ. ```align(s1, s2)```)
* Έναν θετικό αριθμό ίσο με το πλήθος των κενών που πρέπει να μπουν για να γίνει το alignment. Π.χ.:
  * ```align(s1, s3)``` --> 1
  * ```align(s1, s4)``` --> 3
* -1 Αν δεν υπάρχει τρόπος να γίνει το alignment. Π.χ ```align(s1, s5)```

Χρησιμοποιήστε τον παρακάτω πολύ απλό αλγόριθμο alignment:

```

|
```

```

 |
```

Αν το γράμμα που περιέχει το ```k``` δεν είναι το ίδιο με το αντίστοιχο του 1ου string τότε μετακινείστε όλο το  "unaligned" string (δλδ το ```TAT``` στο παράδειγμα) δεξιά πάνω στο 1ο string.  

```

  |
```

